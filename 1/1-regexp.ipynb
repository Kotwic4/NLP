```python
import regex
import os
import collections
```

```python
flatten = lambda l: [item for sublist in l for item in sublist]

def filesNames():
    path = './ustawy'
    absolute_path = os.getcwd() + "\\ustawy\\"
    return [absolute_path + filename for filename in os.listdir(path)]

def getFileTextRaw(filename):
    with open(filename, 'r', encoding="utf8") as content_file:
        return content_file.read()

def getFileText(filename):
    return getFileTextRaw(filename).replace('-\n', '').replace('\n', ' ')

filesTexts = [getFileText(filename) for filename in filesNames()]
```

```python
def getYear(text):
    main = regex.findall(' \d\d\d\d ', text)
    if len(main) > 0:
        return main[0][1:5]
    else:
        return None
```

```python
def findExternalPositions(text, year):
    positions = regex.findall(r'(nr(\d+),poz\.(\d+))', text)
    return [(x[1], x[2], year) for x in positions]
```

```python
def getExternalWithYear(text, defaultYear):
    byYear = regex.findall(r'((z(\d{4})r\.)?((nr\d+,poz\.\d+)(,)?)+(;)?)', text)
    return flatten([findExternalPositions(x[0], x[2] or defaultYear) for x in byYear])
```

```python
def getExternals(text):
    year = getYear(text)
    preText = text.lower().replace(' ', '').replace('i', ',').replace('oraz', ';')
    externals = regex.findall(r'((dz\.u\.)((z(\d{4})r\.)?((nr\d+,poz\.\d+)(,)?)+(;)?)+)', preText)
    return flatten([getExternalWithYear(x[0], year) for x in externals])
```

```python
def pretyForm(text):
    year = text[0][2]
    nr = text[0][0]
    poz = text[0][1]
    counts = text[1]
    return 'year: {}, nr: {}, poz: {}, counts: {}'.format(year, nr, poz, counts)
```

```python
def zad1():
    filesTexts = [getFileText(filename) for filename in filesNames()]
    allExternals = flatten([getExternals(text) for text in filesTexts])
    counter=collections.Counter(allExternals)
    result = [(x, counter[x]) for x in counter.keys()]
    result.sort(key = lambda x: x[1], reverse=True)
    return [pretyForm(x) for x in result]
```

```python
def zad3Help(text):
    options_set = list(set(['ustawa', 'ustawy', 'ustawy', 'ustaw', 'ustawie', 'ustawom', 'ustaw', 'ustawy', 'ustaw', 'ustawami', 'ustawie', 'ustawach', 'ustawo', 'ustawy']))
    p = regex.compile(r"\b\L<options>\b", options=options_set)
    return len(regex.findall(p, text.lower()))

def zad3():
    filesTexts = [getFileText(filename) for filename in filesNames()]
    return sum([zad3Help(text) for text in filesTexts])
```
